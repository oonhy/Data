```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
plt.rc('axes', unicode_minus=False)
```

```python
import pandas as pd

url = 'https://drive.google.com/file/d/1BLIFGyZZGxdMn1hk-RXUtBaE09Sh2lIM/view?usp=drive_link'
file_id = url.split('/')[-2]
download_link = 'https://drive.google.com/uc?id=' + file_id

df_student = pd.read_csv(download_link)

display(df_student)
```

```python
grade_counts = df_student['grade'].value_counts().sort_index()

colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99']

plt.figure(figsize=(10, 6))
plt.bar(
    grade_counts.index,
    grade_counts.values,
    color=colors[:len(grade_counts)],
    edgecolor='black',
    linewidth=1.2
)
plt.title('Frequency of Each Grade', fontsize=18, fontweight='bold', color='darkblue')
plt.xlabel('Grade', fontsize=14, fontweight='bold')
plt.ylabel('Frequency', fontsize=14, fontweight='bold')
plt.xticks(ticks=[1, 2, 3, 4], labels=['1', '2', '3', '4'], fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, grade_counts.values.max() * 1.1)
for i, v in enumerate(grade_counts.values):
    plt.text(grade_counts.index[i], v + 0.5, str(v), ha='center', fontsize=12, color='darkblue')
plt.tight_layout()

plt.show()
```

```python
import seaborn as sns

department_counts = df_student['department'].value_counts()

palette = sns.color_palette('husl', len(department_counts))

plt.figure(figsize=(12, 8))
plt.barh(
    department_counts.index,
    department_counts.values,
    color=palette,
    edgecolor='black',
    linewidth=1.2
)

plt.title('Distribution of Students by Department', fontsize=18, fontweight='bold', color='darkred')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

```python
# Branch

num = 15
if num % 2 == 0:
  print('even number')
else:
  print('odd number')

# nested branches
num = -4
if num > 0:
  if num % 2 == 0:
    print('positive even number')
  else:
    print('positive odd number')
else:
  if num % 2 == 0:
    print('negative even number')
  else:
    print('negative odd number')

# branch with more than 2 cases
age = 22
if age < 13:
  print('child')
elif age >= 13 and age < 20:
  print('teenager')
else:
  print('adult')
```

```python
# Loop

num = 10
result = 1
for i in range(num, 0, -1):
  result = result * i
print(result)

num = 10
result = 1
for i in range(1, num+1, 1):
  result = result * i
print(result)

num = 10
result = 1
while num > 0:
  result = result * num
  num = num - 1
print(result)
```

```python
dummy_data = list()
for i in range(0, 30, 1):
  dummy_data.append('Red')
for i in range(0, 10, 1):
  dummy_data.append('Green')
  dummy_data.append('Blue')
random.shuffle(dummy_data)

df_survey = pd.DataFrame({'Preferred Color': dummy_data})
display(df_survey)
```

```python
color_counts = df_survey['Preferred Color'].value_counts().reset_index()
color_counts.columns = ['Preferred Color', 'Count']

display(color_counts)
```

```python
color_map = {'Red': 'black','Blue':'blue','Green': 'green'}
bar_colors = [color_map[color] for color in color_counts['Preferred Color']]

plt.bar(color_counts['Preferred Color'], color_counts['Count'], color=bar_colors)
plt.title('Preferred Colors by Respondent Count')
plt.xlabel('Preferred Color')
plt.ylabel('Number of Respondents')
plt.show()
```

```python
ordered_colors = ['Blue', 'Red', 'Green']
color_counts_ordered = color_counts.set_index('Preferred Color').reindex(ordered_colors).reset_index()

bar_colors_ordered = [color_map[color] for color in color_counts_ordered['Preferred Color']]

plt.bar(color_counts_ordered['Preferred Color'], color_counts_ordered['Count'], color=bar_colors_ordered)
plt.title('Preferred Colors by Respondent Count')
plt.xlabel('Preferred Color')
plt.ylabel('Number of Respondents')
plt.show()
```

```python
bar_widths_adjusted = [1, 1, 1]
x_positions = [0.5, 1.5, 2.5]

plt.bar(x_positions, color_counts_ordered['Count'], color=ordered_colors, width=bar_widths_adjusted)

plt.xticks(x_positions, color_counts_ordered['Preferred Color'])
plt.title('Preferred Colors by Respondent Count')
plt.xlabel('Preferred Color')
plt.ylabel('Number of Respondents')
plt.show()
```

```python
def stem_leaf_plot(data):
    data = sorted(data)

    stem_leaf = {}
    for num in data:
        stem, leaf = divmod(num, 10)
        if stem not in stem_leaf:
            stem_leaf[stem] = []
        stem_leaf[stem].append(leaf)

    print('Stem | Leaf')
    print('-----|-----')
    for stem in sorted(stem_leaf.keys()):
        leaves = ' '.join(str(leaf) for leaf in stem_leaf[stem])
        print(f'{stem:>4} | {leaves}')


scores = [48, 59, 63, 63, 63, 67, 70, 70, 70, 72, 73, 75, 75,
          80, 81, 82, 83, 84, 87, 88, 88, 90, 91, 95, 95, 95, 96,
          100]

stem_leaf_plot(scores)
```

```python
income_ranges = [(0, 10), (10, 25), (25, 50), (50, 100), (100, 150)]
percentages = [20, 28, 27, 18, 7]

income_midpoints = [(low + high) / 2 for low, high in income_ranges]
income_widths = [high - low for low, high in income_ranges]

plt.bar(
    income_midpoints,
    percentages,
    width=income_widths,
    align='center',
    edgecolor='black',
    color='white',
)

plt.title('Income Distribution Histogram')
plt.xlabel('Income')
plt.ylabel('Percentage')
plt.xticks([0, 10, 25, 50, 100, 150])
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.show()
```

```python
heights = [percent / width for percent, width in zip(percentages, income_widths)]

plt.bar(
    income_midpoints,
    heights,
    width=income_widths,
    align='center',
    edgecolor='black',
    color='white',
)

plt.title('Income Distribution Histogram')
plt.xlabel('Income')
plt.ylabel('Density (Percent per Unit Income)')
plt.xticks([0, 10, 25, 50, 100, 150])
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.show()
```

```python
bar_colors = ['yellow', 'yellow', 'white', 'white', 'white']

plt.bar(
    income_midpoints,
    heights,
    width=income_widths,
    align='center',
    edgecolor='black',
    color=bar_colors,
    alpha=0.9
)

plt.title('Income Distribution Histogram')
plt.xlabel('Income')
plt.ylabel('Density (Percent per Unit Income)')
plt.xticks([0, 10, 25, 50, 100, 150])
plt.grid(axis='y', linestyle='--', alpha=0.6, zorder=0)

plt.show()
```

```python
data = [48, 59, 63, 63, 67, 70, 72, 75, 80, 81, 83, 84, 87, 88, 90, 95, 100]

plt.figure(figsize=(8, 6))
box = plt.boxplot(
    data,
    vert=True,
    patch_artist=True,
    boxprops=dict(facecolor='#FFFFFF', color='#000000', linewidth=1.5),
    medianprops=dict(color='red', linewidth=2),
    zorder=1
)

jitter = np.random.uniform(-0.05, 0.05, size=len(data))
plt.scatter(
    np.ones_like(data) + jitter,
    data,
    color='#2C3E50',
    alpha=0.8,
    s=40,
    label='Data Points',
    zorder=2
)

plt.title('Box Plot', fontsize=16, fontweight='bold')
plt.ylabel('Values', fontsize=12)
plt.xticks([1], ['Data'], fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(loc='upper left', fontsize=10)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

#  
np.random.seed(42)
n_samples = 1000

# 1.  
normal_data = np.random.normal(0, 1, n_samples)

# 2.    ()
skewed_data = np.random.exponential(1, n_samples)

# 3.     (t)
heavy_tail_data = np.random.standard_t(df=3, size=n_samples)

#  
fig, axes = plt.subplots(3, 2, figsize=(12, 15))
fig.suptitle('   Q-Q Plot ', fontsize=16)

# 1. 
axes[0,0].hist(normal_data, bins=30, density=True, alpha=0.7)
axes[0,0].set_title(' ')
stats.probplot(normal_data, dist="norm", plot=axes[0,1])
axes[0,1].set_title(' Q-Q Plot')

# 2.   
axes[1,0].hist(skewed_data, bins=30, density=True, alpha=0.7)
axes[1,0].set_title('  ')
stats.probplot(skewed_data, dist="norm", plot=axes[1,1])
axes[1,1].set_title('  Q-Q Plot')

# 3.   
axes[2,0].hist(heavy_tail_data, bins=30, density=True, alpha=0.7)
axes[2,0].set_title('   ')
stats.probplot(heavy_tail_data, dist="norm", plot=axes[2,1])
axes[2,1].set_title('   Q-Q Plot')

plt.tight_layout()
plt.show()
```
